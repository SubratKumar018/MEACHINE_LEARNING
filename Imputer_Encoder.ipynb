```python
df['Gender'] = lab_enc.fit_transform(df['Gender'])
df
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import make_column_transformer
```

```python
ohe = OneHotEncoder()
si = SimpleImputer()
```

```python
import pandas as pd

df  =  pd.DataFrame({'Salary':[25000,48000,71000,85000,90000,55000],
                    'City':['BNLR','DEL','HYD','BNLR','HYD','BNLR'],
                    'Gender':['M','F','F','F','M','M'],
                    'Exp':[1,3,5,6,9,None]})
df
```

```python
ct = make_column_transformer(
    (ohe,['City','Gender']),
    (si,['Exp']),
    remainder = 'passthrough'
)
```

```python
encoded = pd.DataFrame(ct.fit_transform(df))
encoded
```

```python
encoded = pd.DataFrame(ct.fit_transform(df),columns=['BNLR','DEL','HYD','F','M','EXP','SALARY'])
encoded
```

## OrdinalEncoder

```python
from sklearn.preprocessing import OrdinalEncoder 
```

```python
import pandas as pd

EMP  =  pd.DataFrame({'Salary':[25000,88000,71000,45000],
                    'Position':['SE','Manager','Team Lead','SSE'],
                    'Project':['A','B','C','D']
                    })
EMP
```

```python
ord_enc = OrdinalEncoder(categories=[['SE','SSE','Team Lead','Manager'],['A','B','C','D']])
Encoded_df = ord_enc.fit_transform(EMP[['Position','Project']])
```

# KNNImputer

```python
import pandas as pd

df  =  pd.DataFrame({'Salary':[25000,48000,71000,85000,90000,55000],
                    'City':['BNLR','DEL','HYD','BNLR','HYD','BNLR'],
                    'Gender':['M','F','F','F','M','M'],
                    'Exp':[1,3,5,6,9,None]})
df
```

```python
from sklearn.impute import KNNImputer
```

```python
knnimp = KNNImputer(n_neighbors=3)
knn_imp = pd.DataFrame(knnimp.fit_transform(df[['Salary','Exp']]))
knn_imp
```
